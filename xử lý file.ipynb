```python
path = "Salaries.csv"

f = open(path)
noi_dung = f.readlines()
# print(noi_dung[2])

### xây dựng 1 list có 148645 phần tử
### với từng phần tử của list là 1 dictionary có key là header và value là giá trị của header đó
### ví dụ phần tử đầu tiên sẽ là 1 dictionary:
###### {"Id":"1","EmployeeName":"NATHANIEL FORD","JobTitle":"GENERAL MANAGER-METROPOLITAN TRANSIT AUTHORITY","BasePay":"167411.18",...}
###### {"Id":"2","EmployeeName":"GARY JIMENEZ","JobTitle":"CAPTAIN III (POLICE DEPARTMENT)","BasePay":"155966.02",...}

L_data = []
header = noi_dung[0].strip().split(",")

def check(L_chuoi):
    L_kq = []
    c = False
    for i in range(len(L_chuoi)):
        if c:
            c = False
            continue
        if '"' not in L_chuoi[i]:
            L_kq.append(L_chuoi[i])
        else:
            if '"' not in L_chuoi[i+1]:
                return None
            L_kq.append(L_chuoi[i].replace('"',"") + "," + L_chuoi[i+1].replace('"',""))
            c = True
    return L_kq

for i in range(1,len(noi_dung)):
    noi_dung_1 = check(noi_dung[i].strip().split(","))
    if noi_dung_1 == None:
        continue
    D = {}
    for i in range(len(header)):
        D[header[i]] = noi_dung_1[i]
    L_data.append(D)

### in ra tên những người có lương cơ bản > 100000
for data in L_data:
    if data['BasePay'] != "" and float(data['BasePay']) > 100000:
        print(data["EmployeeName"])
```
