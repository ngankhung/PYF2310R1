```python
from selenium import webdriver
import time

driver = webdriver.Chrome()

driver.get("https://tiki.vn/")

time.sleep(1)

inp = driver.find_element("class name","styles__InputRevamp-sc-6cbqeh-2")
inp.send_keys("balo")

but = driver.find_element("class name","styles__ButtonRevamp-sc-6cbqeh-3")
but.click()

time.sleep(2)

data_element = driver.find_element("class name","CatalogProducts__Wrapper-sc-1hmhz3p-0")
children = data_element.find_elements("xpath", '*')
for child in children:
    try:
        name = child.find_element("class name","name").get_attribute("innerHTML").replace("<h3>","").replace("</h3>","")
        print(name)
        price = child.find_element("class name","price-discount__price").get_attribute("innerHTML").replace("<sup>₫</sup>","")
        print(price)
    except:
        pass
```
