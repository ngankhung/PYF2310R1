```python
from tkinter import*

calculator = Tk()
calculator.geometry("360x428")
calculator.resizable(0,0)
calculator.title("Calculator")

default_font = ("Arial",12,)
display_font = ("Arial",24,"bold")
display_text = StringVar()
expression = ""

def click_button(key):
    global expression
    expression += key
    display_text.set(expression)
    
def clear_button():
    global expression
    expression = ""
    display_text.set(expression)

def undo_button():
    global expression
    current_expression = expression[:-1]
    expression = current_expression
    display_text.set(current_expression)

def calculate():
    global expression
    try:
        if expression != "":
            if "\u00F7" in expression:
                result = eval(expression.replace("\u00F7", "/"))
            else:
                result = eval(expression)
            result = int(result)
            expression = str(result)
            display_text.set(expression)
    except:
        display_text.set("Invalid")

def percent_button():
    global expression
    try:
        result = eval(expression)
        result_percent = result/100
        expression = str(result_percent)
        display_text.set(expression)
    except:
        display_text.set("Invalid")

def power_button():
    click_button("**")

def fraction_button():
    global expression
    try:
        result = eval(expression)
        result_divided = 1/result
        expression = str(result_divided)
        display_text.set(expression)
    except:
        display_text.set("Invalid")

def square_button():
    global expression
    try:
        result = eval(expression)
        result_squared = result ** 2
        expression = str(result_squared)
        display_text.set(expression)
    except:
        display_text.set("Invalid")

def square_root_button():
    global expression
    try:
        result = eval(expression)
        result_squared_root = result ** 0.5
        expression = str(result_squared_root)
        display_text.set(expression)
    except:
        display_text.set("Invalid")

def sign_change():
    global expression
    if expression[0] =="-":
        result = expression[1:]
    else:
        result = "-" + expression
    expression = result
    display_text.set(result)

display = Label(calculator, font=display_font, textvariable=display_text, height=2, anchor="e", justify="right")
display.grid(row=0,columnspan=6,sticky=N+E+W+S)

zero = Button(calculator, text="0", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("0"))
zero.grid(column=2, row=7)
one = Button(calculator, text="1", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("1"))
one.grid(column=1, row=6)
two = Button(calculator, text="2", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("2"))
two.grid(column=2, row=6)
three = Button(calculator, text="3", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("3"))
three.grid(column=3, row=6)
four = Button(calculator, text="4", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("4"))
four.grid(column=1, row=5)
five = Button(calculator, text="5", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("5"))
five.grid(column=2, row=5)
six = Button(calculator, text="6", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("6"))
six.grid(column=3, row=5)
seven = Button(calculator, text="7", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("7"))
seven.grid(column=1, row=4)
eight = Button(calculator, text="8", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("8"))
eight.grid(column=2, row=4)
nine = Button(calculator, text="9", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("9"))
nine.grid(column=3, row=4)

plus = Button(calculator, text="+", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("+"))
plus.grid(column=4, row=6)
minus = Button(calculator, text="-", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("-"))
minus.grid(column=4, row=5)
multiply = Button(calculator, text="*", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("*"))
multiply.grid(column=4, row=4)
divide = Button(calculator, text="\u00F7", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("\u00F7"))
divide.grid(column=4, row=3)

decimal = Button(calculator, text=".", font=default_font, width=8, height=2, padx=5, pady=5, command=lambda:click_button("."))
decimal.grid(column=3, row=7)
percent = Button(calculator, text="%", font=default_font, width=8, height=2, padx=5, pady=5, command=percent_button)
percent.grid(column=1, row=2)
fraction = Button(calculator, text="\u215Fx", font=default_font, width=8, height=2, padx=5, pady=5, command=fraction_button)
fraction.grid(column=1, row=3)
square = Button(calculator, text="xÂ²", font=default_font, width=8, height=2, padx=5, pady=5, command=square_button)
square.grid(column=2, row=3)
square_root = Button(calculator, text="\u221Ax", font=default_font, width=8, height=2, padx=5, pady=5, command=square_root_button)
square_root.grid(column=3, row=3)
minus_sign = Button(calculator, text="+/-", font=default_font, width=8, height=2, padx=5, pady=5, command=sign_change)
minus_sign.grid(column=1, row=7)

equal = Button(calculator, text="=", font=default_font, width=8, height=2, padx=5, pady=5, command=calculate)
equal.grid(column=4, row=7)
clear = Button(calculator, text="C", font=default_font, width=8, height=2, padx=5, pady=5, command=clear_button)
clear.grid(column=3, row=2)
power_button = Button(calculator, text="x^y", font=default_font, width=8, height=2, padx=5, pady=5, command=power_button)
power_button.grid(column=2, row=2)
undo = Button(calculator, text="\u232B", font=default_font, width=8, height=2, padx=5, pady=5, command=undo_button)
undo.grid(column=4, row=2)

calculator.mainloop()
```
